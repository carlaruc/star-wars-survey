# Star wars Survey results

While waiting for Star Wars: The Force Awakens to come out, the team at FiveThirtyEight became interested in answering some questions about Star Wars fans. In particular, they wondered: does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?

For this project we are going to clean and explore the dataset in Jupyter notebooks.

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding = "ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

## Cleaning and Mapping Yes/No Columns

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts()
```

```python
yes_no = {"Yes": True, "No": False}

for col in ["Have you seen any of the 6 films in the Star Wars franchise?", "Do you consider yourself to be a fan of the Star Wars film franchise?"]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

## Cleaning and Mapping Checkbox Columns

```python
import numpy as np
def yes_no(x):
    if isinstance(x,str):
        return True
    else:
        return False

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].apply(yes_no)
```

```python
star_wars[star_wars.columns[5]].value_counts
```

```python
print(star_wars)
```

```python
renaming_columns = {}
c = 1
for col in star_wars.columns[3:9]:
    renaming_columns[col] = f"seen_{c}"
    c += 1

star_wars.rename(columns=renaming_columns, inplace=True)
```

```python
print(star_wars.columns)
```

# Cleaning the Ranking Columns

```python
print(star_wars.columns)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars.dtypes
```

```python
column_names = {}
c=1

for col in star_wars.columns[9:15]:
    column_names[col] = f"ranking_{c}"
    c+=1

star_wars.rename(columns=column_names, inplace=True)
```
